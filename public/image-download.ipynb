```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse


def download_images(url, base_folder):
    """Download all images from a given URL and save them to the specified path."""

    # Fetch the webpage
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an error for bad status codes
    except requests.RequestException as e:
        print(f"Failed to fetch {url}: {e}")
        return

    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all image tags
    img_tags = soup.find_all('img')

    for img in img_tags:
        # Get the absolute URL of the image
        img_url = urljoin(url, img.get('src', ''))

        # Extract the folder name from the img src
        img_src = img.get("src", "")
        parsed_src = urlparse("/".join(img_src.split("/")[2:-1]))
        folder_name = os.path.join(base_folder, parsed_src.path.strip("/"))

        if not folder_name:
            folder_name = "uncategorized"  # Default folder if src path is empty

        # Create the folder if it doesn't exist
        if not os.path.exists(folder_name):
            os.makedirs(folder_name)

        try:
            # Download the image
            img_data = requests.get(img_url).content
            img_name = os.path.join(folder_name, img_url.split("/")[-1])

            with open(img_name, "wb") as img_file:
                img_file.write(img_data)
                print(f"Downloaded {img_name} to {folder_name}")
        except requests.RequestException as e:
            print(f"Failed to download image {img_url}: {e}")


def scrape_site(base_url, start_route, target_route, visited=None, first=True):
    """Scrape all pages under a target route, starting from a specific route."""
    if visited is None:
        visited = set()

        # Construct the full URL for the starting route

    full_url = base_url + start_route

    # Avoid revisiting the same page
    if full_url in visited:
        return
    visited.add(full_url)

    # Fetch the webpage
    try:
        response = requests.get(full_url)
        response.raise_for_status()
    except requests.RequestException as e:
        print(f"Failed to fetch {full_url}: {e}")
        return

    if first:
        soup = BeautifulSoup(html, "html.parser")
    else:
        soup = BeautifulSoup(response.text, 'html.parser')

    # If the current page is under the target route, download its images
    if not first and start_route.startswith(target_route):
        # Download images from the current page
        print(f"Scraping {full_url}")
        download_images(full_url, "pages2")

    # Find all links on the current page
    for link in soup.find_all('a', href=True):
        href = link['href']
        # Ensure the link is under the target route
        if href.startswith(target_route) or href.startswith('/' + target_route):
            # Recursively scrape the sub-route
            scrape_site(base_url, href, target_route, visited, False)

if __name__ == "__main__":
    base_url = "https://www.tu-chemnitz.de/informatik/theoretische-informatik/TI-2/"
    target_route = 'lecture-notes' 
    start_route = 'vorlesungsskript.html' 
    scrape_site(base_url, start_route, target_route)
```
